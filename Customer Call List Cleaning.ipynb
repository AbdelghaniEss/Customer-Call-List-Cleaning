# Data Cleaning and EDA using Panda

Let's import libraries

```python
import pandas as pd
```

I was working in a cloud based environment so that's why I uploaded the dataset directly 

```python
df = pd.read_csv('Customer Call List.csv')
df
```

We start by removing duplicate rows

```python
df = df.drop_duplicates()
df
```

Next we remove irrelevant columns for EDA that we will perform later

```python
df = df.drop('Not_Useful_Column', axis=1)
df
```

After checking the columns we can see that the Last_Name column has a few special characters that should be removed

```python
df['Last_Name'] = df['Last_Name'].str.strip('/._')
df
```

Now we can see that the Phone_Number column's values are not consistent and don't follow one similar format so we need to rectify that

First, let's remove the special characters

```python
df['Phone_Number'] = df['Phone_Number'].str.replace('[^a-zA-Z0-9]', '', regex=True)
df
```

Next we format the column so that it follows one consitent format for all rows with phone numbers data

```python
df['Phone_Number'] = df['Phone_Number'].apply( lambda x: str(x)[0:3] + '-' + str(x)[3:6] + '-' + str(x)[6:10])
df
```

Finally we replace the Na-- and nan-- values with empty space since our agents won't need anyway when they will call our customers

```python
df['Phone_Number'] = df['Phone_Number'].str.replace('Na--|nan--', '', regex=True)
df
```

Now looking at the Address column we have a few rows with the State and zip code, let's split this information into 3 seperated columns

```python
df[['Street_Address', 'State', 'Zip_Code']] = df['Address'].str.split(',', n=2, expand=True)
df
```

In the Paying Customer column we can see that a few Yes and No values were entered as Y or N, we will rectify this as well

```python
df['Paying Customer'] = df['Paying Customer'].apply( lambda x: 'Yes' if x == 'Y' else ('No' if x == 'N' else x))
df
```

We do the same for the Do_Not_Contact column

```python
df['Do_Not_Contact'] = df['Do_Not_Contact'].apply( lambda x: 'Yes' if x == 'Y' else ('No' if x == 'N' else x))
df
```

The last step in cleaning columns is to fill all the Nan and None values with an empty space

```python
df = df.fillna('')
df
```

We see that we still have a few N/a values these should replaced directly

```python
df = df.replace('N/a', '')
df
```

We will now filter our DataFrame to remove customers who requested not to be contacted, and we will retain only the records with valid phone numbers. This step is crucial to ensure we provide our customer service team with a clean and accurate dataset for their call outreach.

```python
filtered_df = df[(df['Do_Not_Contact'] != 'Yes') & (df['Phone_Number']!= '')]
filtered_df
```

Now let's just reset the index to help us have a cleared idea of how many customers are there

```python
filtered_df = filtered_df.reset_index(drop=True)
filtered_df
```
